실생활 예시 코드1: Python 딕셔너리를 사용하여 친구 관계(네트워크 그래프)를 표현하고, 두 사람이 직접적인 친구인지 확인하는 함수를 작성합니다.

```python
# 과학자들의 협력 관계를 네트워크로 표현 (아인슈타인은 보어, 슈뢰딩거와 토론함)
scientist_network = {
    "Einstein": ["Bohr", "Schrodinger"],
    "Bohr": ["Einstein", "Heisenberg"],
    "Schrodinger": ["Einstein"],
    "Heisenberg": ["Bohr"]
}

def are_they_collaborators(network, person1, person2):
    # person1의 협력자 목록에 person2가 있는지 확인
    if person1 in network and person2 in network[person1]:
        return True
    return False

p1, p2 = "Einstein", "Bohr"
if are_they_collaborators(scientist_network, p1, p2):
    print(f"{p1}와(과) {p2}는 직접적인 협력 관계입니다.")
else:
    print(f"{p1}와(과) {p2}는 직접적인 협력 관계가 아닙니다.")
```

수학/과학 예시 코드: 원주율(\pi)의 긴 소수점 값을 일정한 크기의 패킷으로 나누고, 각 패킷에 헤더(순서, 데이터 길이)를 붙이는 함수를 작성합니다.

```python
import math

def create_packets(long_data_string, packet_size):
    packets = []
    # 긴 데이터를 packet_size에 맞춰 자르기
    for i in range(0, len(long_data_string), packet_size):
        payload = long_data_string[i:i+packet_size]
        # 패킷 헤더: (순서 번호, 데이터 길이) + 실제 데이터(payload)
        packet = {
            "sequence": (i // packet_size) + 1,
            "length": len(payload),
            "payload": payload
        }
        packets.append(packet)
    return packets

# 원주율 소수점 이하 50자리를 데이터로 사용
pi_data = str(math.pi)[2:52]
print("원본 데이터:", pi_data)

# 10글자씩 패킷으로 나눔
packet_list = create_packets(pi_data, 10)
print("\n생성된 패킷:")
for p in packet_list:
    print(p)
```

실생활 예시 코드: 문자열을 입력받아, 그것이 유효한 IPv4 주소 형식인지 검사하는 함수를 작성합니다.

```python
def is_valid_ipv4(ip_string):
    parts = ip_string.split('.')
    if len(parts) != 4:
        return False
    for part in parts:
        if not part.isdigit(): # 숫자가 아닌 문자가 있으면 False
            return False
        num = int(part)
        if not 0 <= num <= 255: # 숫자의 범위가 0~255가 아니면 False
            return False
    return True

ip1 = "192.168.0.1"
ip2 = "256.0.0.1" # 255보다 큰 숫자가 있어 유효하지 않음
print(f"'{ip1}'는 유효한 IPv4 주소인가? {is_valid_ipv4(ip1)}")
print(f"'{ip2}'는 유효한 IPv4 주소인가? {is_valid_ipv4(ip2)}")
```

실생활 예시 코드: 두 개의 IP 주소가 같은 LAN에 속하는지 (앞의 세 자리가 같은지) 간단하게 확인하는 함수를 작성합니다.

```python
def are_in_same_lan_simple(ip1, ip2):
    # IP 주소의 앞 세 부분(네트워크 ID)이 같은지 비교
    ip1_net_id = ".".join(ip1.split('.')[:3])
    ip2_net_id = ".".join(ip2.split('.')[:3])

    if ip1_net_id == ip2_net_id:
        return True
    return False

my_pc = "192.168.1.10"
my_friend_pc = "192.168.1.25"
school_server = "210.117.123.5"

print(f"{my_pc}와(과) {my_friend_pc}는 같은 LAN에 있는가? {are_in_same_lan_simple(my_pc, my_friend_pc)}")
print(f"{my_pc}와(과) {school_server}는 같은 LAN에 있는가? {are_in_same_lan_simple(my_pc, school_server)}")
```

IP 주소 유효성 검사기
사용자로부터 문자열을 입력받아, 해당 문자열이 유효한 IPv4 주소 형식인지 판별하는 프로그램을 작성하세요. (조건: 4개의 부분으로 나눠, 각 부분은 0-255 사이의 숫자)

```python
def validate_ipv4(ip_string):
    """입력된 문자열이 유효한 IPv4 주소 형식인지 판별합니다."""
    parts = ip_string.split('.')
    if len(parts) != 4:
        return False
    for part in parts:
        if not part.isdigit():
            return False
        num = int(part)
        if not 0 <= num <= 255:
            return False
    return True

# 테스트
user_input = input("IPv4 주소를 입력하세요: ")
if validate_ipv4(user_input):
    print(f"'{user_input}'는 유효한 IPv4 주소 형식입니다.")
else:
    print(f"'{user_input}'는 유효하지 않은 IPv4 주소 형식입니다.")
```

간단한 패킷 생성기
문자열 데이터와 패킷당 최대 크기를 입력받아, 데이터를 여러 개의 패킷(딕셔너리)으로 나누어 리스트에 담아 반환하는 함수를 작성하세요. 각 패킷에는 순서 번호와 데이터 조각(payload)이 포함되어야 합니다.

```python
def create_packets(data, max_size):
    """긴 데이터를 주어진 크기의 패킷으로 분할합니다."""
    packets = []
    packet_number = 1
    for i in range(0, len(data), max_size):
        payload = data[i : i + max_size]
        packet = {
            'sequence': packet_number,
            'payload': payload
        }
        packets.append(packet)
        packet_number += 1
    return packets

# 테스트
long_message = "The best way to predict the future is to invent it. - Alan Kay"
packet_list = create_packets(long_message, 20) # 최대 20글자씩 자름
for packet in packet_list:
    print(packet)
```

패킷 조립기
위 2번 문제에서 생성된 패킷 리스트를 받아, 순서 번호에 맞게 조립하여 원본 데이터를 복원하는 함수를 작성하세요.

```python
import random

def assemble_packets(packet_list):
    """분할된 패킷 리스트를 순서에 맞게 조립하여 원본 데이터로 복원합니다."""
    # 순서 번호(sequence)를 기준으로 패킷 리스트를 정렬합니다.
    # 혹시 패킷이 뒤죽박죽 도착하는 상황을 시뮬레이션합니다.
    sorted_packets = sorted(packet_list, key=lambda p: p['sequence'])

    original_data = ""
    for packet in sorted_packets:
        original_data += packet['payload']

    return original_data

# 테스트 (위 2번 예제의 packet_list를 사용)
print("\n뒤죽박죽 도착한 패킷 순서 (시뮬레이션):")
random.shuffle(packet_list) # 패킷 순서를 일부러 섞어봄
for p in packet_list:
    print(p['sequence'], end=' ')

restored_message = assemble_packets(packet_list)
print("\n\n복원된 메시지:")
print(restored_message)
```

로컬 네트워크(LAN) 확인기

두 개의 IPv4 주소 문자열을 입력받아, 두 주소가 같은 C 클래스 LAN에 속하는지(즉, IP 주소의 앞 세 부분이 동일한지) 판별하는 함수를 작성하세요.

```python
# IP별로 열려있는 포트와 서비스 정보를 담은 딕셔너리
server_ports = {
    "210.117.123.10": {
        80: "HTTP (Web Server)",
        443: "HTTPS (Secure Web Server)",
        21: "FTP (File Transfer)"
    },
    "210.117.123.11": {
        22: "SSH (Secure Shell)",
        3306: "MySQL (Database)"
    }
}

def check_service_on_port(ip, port):
    """주어진 IP와 포트 번호에서 실행 중인 서비스를 확인합니다."""
    # 해당 IP 정보가 있는지 먼저 확인
    if ip in server_ports:
        server = server_ports[ip]
        # 해당 서버에 요청한 포트 정보가 있는지 확인
        service = server.get(port)
        if service:
            print(f"서버 {ip}의 {port}번 포트에서는 '{service}' 서비스가 실행 중입니다.")
        else:
            print(f"서버 {ip}의 {port}번 포트는 열려있지 않거나 서비스 정보가 없습니다.")
    else:
        print(f"서버({ip}) 정보를 찾을 수 없습니다.")

# 테스트
check_service_on_port("210.117.123.10", 80)
check_service_on_port("210.117.123.10", 25) # 25번 포트는 목록에 없음
check_service_on_port("210.117.123.11", 22)
check_service_on_port("127.0.0.1", 8080) # IP가 목록에 없음
```
